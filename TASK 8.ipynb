```python
import pandas as pd
```

```python
df= pd.read_csv("retail_store_sales.csv")
```

```python
missingpercent = df.isnull().mean() * 100
missingpercent.sort_values(ascending=False)
```

```python
df["Price Per Unit"].isna().sum()
```

```python
df["Price Per Unit"] = pd.to_numeric(df["Price Per Unit"], errors="coerce")
```

```python
df["Price Per Unit"].value_counts()
```

```python
df["Total Spent"] = df["Total Spent"].ffill()
```

```python
df["Total Spent"].value_counts()
```

```python
df["Price Per Unit"] = df["Price Per Unit"].fillna(df["Price Per Unit"].mean())
```

```python
df["Quantity"] = df["Quantity"].mode()
```

```python
Q1 = df["Price Per Unit"].quantile(0.25)
Q3 = df["Price Per Unit"].quantile(0.75)

IQR = Q3 - Q1
print("Price Per Unit IQR:", IQR)

lower = Q1 - 1.5 * IQR
print("Price Per Unit lower:", lower)

upper = Q3 + 1.5 * IQR
print("Price Per Unit upper:", upper)

outliers = ((df["Price Per Unit"] < lower) | (df["Price Per Unit"] > upper)).sum()
print("Price Per Unit outliers:", outliers)
```

```python
Q1 = df["Quantity"].quantile(0.25)
Q3 = df["Quantity"].quantile(0.75)

IQR = Q3 - Q1
print("Quantity IQR:", IQR)

lower = Q1 - 1.5 * IQR
print("Quantity lower:", lower)

upper = Q3 + 1.5 * IQR
print("Quantity upper:", upper)

outliers = ((df["Quantity"] < lower) | (df["Quantity"] > upper)).sum()
print("Quantity outliers:", outliers)
```

```python
Q1 = df["Total Spent"].quantile(0.25)
Q3 = df["Total Spent"].quantile(0.75)

IQR = Q3 - Q1
print("Total Spent IQR:", IQR)

lower = Q1 - 1.5 * IQR
print("Total Spent lower:", lower)

upper = Q3 + 1.5 * IQR
print("Total Spent upper:", upper)

outliers = ((df["Total Spent"] < lower) | (df["Total Spent"] > upper)).sum()
print("Total Spent outliers:", outliers)
```
