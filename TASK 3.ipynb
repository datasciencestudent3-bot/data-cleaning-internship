```python
df= pd.read_csv("dirty_cafe_sales.csv", na_values=["error","unknown","N/A"])
```

```python
df["Transaction Date"] = df["Transaction Date"].ffill()
```

```python
(df.isnull().mean() * 100)
```

```python
df["Transaction Date"].head()
```

```python
df["Date_raw"] = df["Transaction Date"]

df["Date"] = pd.to_datetime(df["Date_raw"], errors="coerce", dayfirst=True)

mask = df["Date"].isna()
df.loc[mask, "Date"] = pd.to_datetime(df.loc[mask, "Date_raw"], errors="coerce")

df["Date"] = df["Date"].dt.strftime("%d-%m-%Y")
```
